```python
import numpy as np
import pandas as pd
```

```python
V = np.zeros((3,4))
```

```python
lock = [(1,1)]
gamma = 0.9
V[1,1] = -100
V[0,3] = 1
V[1,3] = -1
V
```

```python
#Hàm tìm kiếm vị trí có thể đi tại một ô khi biết tọa độ x,y của ô đó
def find_move(x,y, shape_v):
  up = (x-1, y)
  down = (x+1, y)
  left = (x, y-1)
  right = (x, y+1)

  list_move = [up, down, left, right]

  true_list_move = []
  for move in list_move:
    for stt,i in enumerate(move):
      if i < 0:
        break
      elif i >= shape_v[stt]:
        break
      if stt:
        true_list_move.append(move)
  for l in lock:
    if l in true_list_move:
      true_list_move.remove(l)
  return true_list_move
```

```python
def allMoves(V):
  results = []
  for i in range(V.shape[0]):
    for j in range(V.shape[1]):
      obj = {(i, j): find_move(i, j, V.shape)}
      results.append(obj)

  return results;

def addValue(value):
    allVal = []

    for val in value:
        allVal.append(val)

    return allVal
```

```python
def updateValue(value, gamma, V):
    listValue = addValue(value)
    maxValue = 0

    if len(value) > 1:
        sumVal = len(value) - 1
    else:
        sumVal = len(value)

    result = []

    for i in range(len(listValue)):
        sum = 0
        x_i, y_i = listValue[i]

        for j in range(len(listValue)):
            x_j, y_j = listValue[j]
            if j == i:
                sum += 0.8 * gamma * V[x_i][y_i]
            else:
                sum += (0.2 / sumVal) * gamma * V[x_j][y_j]
        result.append(sum)

    maxValue = max(result)

    return maxValue
```

```python
listMove = allMoves(V)
for __ in range(20):
  for i in listMove:
    for _ in i:
      if(_ == (0,3) or _ == (1,3) or _ in lock):
         continue
      value = i[_]
      V[_] = updateValue(value, gamma, V)
  print(f'lần lặp: {__+1}')
  display(pd.DataFrame(V))
```
